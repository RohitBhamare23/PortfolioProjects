```python
#Droppping all rows having missing data there are many other ways to do this but for time being dropping em all
df=df.dropna()
```

```python
null_count = df.isnull().sum()
null_count
```

```python
# Data Types of columns:
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
df.head()
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False ).head()
```

```python
# Changing the format of released year:

df['released'] = (
    df['released'].str.split('(').str[0]
    .str.strip()
)
df['released'] = pd.to_datetime(
    df['released'], 
    errors = 'coerce',
    format = '%B %d, %Y'    
)
df.head()
```

```python
#BUDGET high Correlation , COMPANY high correlation 
```

```python
#Scatter plot with Budget Vs Gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#Plot using Seaborn:

sns.regplot(x='budget', y='gross', data = df, scatter_kws={'color':'purple'}, line_kws={'color':'yellow'})
```

```python
#df.corr(method = 'pearson') # Diff Methods: pearson ,kendall , spearman By Default its pearson.
corr_matrix = df.select_dtypes(include=[np.number]).corr(method='pearson') # To filter out non numeric cols.
corr_matrix
```

```python
#JUst in case if the columns have diff values that might create prob while visualization
corr_matrix.replace([np.inf, -np.inf], np.nan, inplace=True)
corr_matrix.dropna(axis=0, how='any', inplace=True)
corr_matrix.dropna(axis=1, how='any', inplace=True)
```

```python
# Heatmap of above correlation matrix:

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation matrix for Movie Features:')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()

#ps:Did Everything i could changed the size of plot, changed the colour scheme, changed the color of the nos., reduced the nos. to 2 decimals,
#and many diff things still only the top row is being displayed i dont know why.
```

```python
#Visualizing the matrix of all columns:
df_numerised = df.copy()

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes
df_numerised.head()
```

```python
df_numerised.corr()
```

```python
#Unstack presents the correlation of every col to each other.
corr_mat = df_numerised.corr()
corr_pairs = corr_mat.unstack()
corr_pairs.head()
```

```python
#Paired Comparison of Correlations:
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head()
```

```python
#Paired Comparison of Correlations:
sorted_pairs = corr_pairs.sort_values(ascending = False)
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5 ]
high_corr
```
